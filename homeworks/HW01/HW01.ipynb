```python
import numpy as np

# Создание массивов
a = np.array([[1, 2, 3], [2, 3, 4]])
b0 = np.arange(0, 100, 5).reshape(-1, 5)
b = np.arange(0, 100, 5).reshape(5, -1)
c = np.random.random(10)

# shape и dtype
print(a.shape)
print(a.dtype)

# Индексирование и срезы
print(b[:])
print()
print(b[3][-1]) # тоже самое что и:
print(b[(3, -1)]) # тоже самое что и:
print(b[3, -1])
print()
print(b[[2, 3]])
print(b[[2, 3], 2])
print()
print(b[:, 2]) # третий столбец
print()
print(b[1:4, : : ])

# операции с массивами
print(b/2)
print(b*2) # умножение массива на число
print(b.dot(b0)) # умножение массива на массив
print(b.mean()) # среднее значение массива
print()
print(b.max(axis=0)) # минимум по строкам для каждого столбца
print(b.max(axis=1)) # минимум по столбцам для каждой строки
```
